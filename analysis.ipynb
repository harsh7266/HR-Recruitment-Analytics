```python
# ==========================================
# STEP 1: IMPORT LIBRARIES (Matching the JD)
# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================
# STEP 2: LOAD THE DATA
# ==========================================
# This reads the 'Raw_Data' sheet from the Excel file I gave you
df = pd.read_excel('QuickHyre_Recruitment_Data.xlsx', sheet_name='Raw_Data')

# Display the first 5 rows to verify it loaded correctly
print("Data Successfully Loaded. Here is a preview:")
display(df.head())

# ==========================================
# STEP 3: DATA TRANSFORMATION & VALIDATION
# ==========================================
# The JD asks for Data Cleaning and Transformation. 
# Let's check for missing values and summarize the data.

print("\n--- Checking for Missing Values ---")
print(df.isnull().sum())

# Create a new specific dataset focusing ONLY on Hired candidates
# This is a 'Transformation' step to prepare for specific analysis
hired_df = df[df['Current_Status'] == 'Hired']

# Calculate the overall Time-to-Hire metric
avg_time = hired_df['Processing_Time_Days'].mean()
print(f"\nAverage Time-to-Hire across all departments: {avg_time:.1f} days")

# ==========================================
# STEP 4: EXPLORATORY DATA ANALYSIS (EDA)
# ==========================================
# The JD asks for EDA using Matplotlib. Let's recreate the Time-to-Hire chart.

# Set the visual style
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(10, 6))

# Create a boxplot to see the distribution of hiring times per department
sns.boxplot(data=hired_df, x='Department', y='Processing_Time_Days', palette='coolwarm')

# Add titles and labels to make it professional
plt.title('Time-to-Hire Distribution by Department', fontsize=16, fontweight='bold')
plt.ylabel('Days to Hire', fontsize=12)
plt.xlabel('Department', fontsize=12)


plt.tight_layout()
plt.show()
```
